```python
import pickle

import numpy as np
import pandas as pd
import seaborn as sns

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.utils.data.sampler import SubsetRandomSampler
```

```python
# Read pickle file
df = pd.read_pickle('../sim/sweep/data1/f0.9/pd-data.pickle')
df.describe()
# df.head()
# df.tail()
```

```python
class SNN(nn.Module):
    
    def __init__(self, input_size, hidden_size, output_size, afcn):
        super(SNN, self).__init__()
        
        self.afcn = afcn
        
        # fc -> fully connected
        self.fc1 = nn.Linear(input_size, hidden_size)
        
        if afcn == 'ReLu':
            self.h_activation = nn.ReLU()
        elif afcn == 'Sigmoid':
            self.h_activation = nn.Sigmoid()
        elif afcn == 'Tanh':
            self.h_activation = nn.Tanh()
        else:
            print('ERROR: invalid activation function', afcn)
        
        self.fc2 = nn.Linear(hidden_size, output_size)
        
    def forward(self, x):
        out = self.fc1(x)
        out = self.h_activation(out)
        out = self.fc2(out)
        return out
# class SNN(nn.Module):
    
#     def __init__(self, input_size, hidden_size1, hidden_size2, output_size, afcn):
#         super(SNN, self).__init__()
        
#         self.afcn = afcn
        
#         self.fc1 = nn.Linear(  input_size, hidden_size1)
#         self.fc2 = nn.Linear(hidden_size1, hidden_size2)
#         self.fc3 = nn.Linear(hidden_size2,  output_size)
        
#         if afcn == 'ReLu':
#             self.squash = nn.ReLU()
#         elif afcn == 'Sigmoid':
#             self.squash = nn.Sigmoid()
#         elif afcn == 'Tanh':
#             self.squash = nn.Tanh()
#         else:
#             print('ERROR: invalid activation function', afcn)
        
        
#     def forward(self, x):
#         out = self.squash(self.fc1(x))
#         out = self.squash(self.fc2(out))
#         out = self.fc3(out)
#         return out
```

```python
class UGVDataset(Dataset):
    
    def __init__(self, x, y):
        
        assert(len(x) == len(y))
        self.len = len(x)
        
        self.x = torch.tensor(x, dtype=torch.float)
        self.y = torch.tensor(y, dtype=torch.float)
        
        self.xsize = self.x.shape[1]
        self.ysize = self.y.shape[1] if len(self.y.shape) > 1 else 1
        
    def __getitem__(self, index):
        return (self.x[index], self.y[index])
    
    def __len__(self):
        return self.len
    
    def input_size(self):
        return self.xsize
    
    def output_size(self):
        return self.ysize
```

```python
# Parameters

network_output = 'dlon'

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print('Device is', device)

hidden_size1 = 4      # 
hidden_size2 = 20      # 
learning_rate = 0.009   # 
batch_size = 128        # 
num_epochs = 100          # 
activation_fcn = 'ReLu' # 

train_pct = 0.8
do_shuffle = True
random_seed = 842
```

```python
# Load the dataset
x = df[['left', 'right', 'strut']].values
y = df[network_output].values
dataset = UGVDataset(x, y)

print('Training:', network_output)

# Create training and testing samplers
dataset_size = len(dataset)
all_indices = list(range(dataset_size))

if do_shuffle:
    np.random.seed(random_seed)
    np.random.shuffle(all_indices)

split_index = int(np.floor(train_pct * dataset_size))

# train_indices = all_indices[:split_index]
train_indices = all_indices
train_sampler = SubsetRandomSampler(train_indices)

test_indices = all_indices[split_index:]
test_sampler = SubsetRandomSampler(test_indices)

# Create dataloaders
train_loader = DataLoader(dataset=dataset, batch_size=batch_size, sampler=train_sampler)
test_loader = DataLoader(dataset=UGVDataset(x,y), batch_size=batch_size, sampler=test_sampler)
```

```python
input_size = dataset.input_size()
output_size = dataset.output_size()

# Neural network
# model = SNN(input_size, hidden_size1, hidden_size2, output_size, activation_fcn)
model = SNN(input_size, hidden_size1, output_size, activation_fcn)

# Optimization function (loss)
# criterion = nn.CrossEntropyLoss() # <-- for classification
criterion = nn.MSELoss()

# Optimization algorithm
# optimizer = optim.Adam(model.parameters(), lr=learning_rate)
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# Training

num_steps = len(train_loader)
print_step = 1

model.to(device)

losses = []

for epoch in range(num_epochs):
    
    running_loss = 0.0
    
    for i, (x, y) in enumerate(train_loader):

        # Move tensors to the configured device
        x = x.reshape(-1, input_size).to(device)
        y = y.to(device)
        
        # Forward pass
        y_hat = model(x)        
        loss = criterion(y_hat, y)
        
        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        i1 = i + 1
        e1 = epoch + 1
        running_loss += loss.item()
        
        if i1 % print_step == 0:
            l = running_loss / print_step
            losses.append(l)
            running_loss = 0.0
            print(f'Epoch: [{e1}/{num_epochs}], Step: [{i1}/{num_steps}], Loss: {l:.4f}')

print('Finished training')
```

```python
sns.lineplot(list(range(len(losses))), losses);
```

```python
# Testing

# In test phase, we don't need to compute gradients (for memory efficiency)
with torch.no_grad():
    
    total_loss = 0
    
    for x, y in test_loader:
    
        x = x.reshape(-1, input_size).to(device)
        y = y.to(device)
        y_hat = model(x)
        total_loss += criterion(y_hat, y)
        
        print('input    ', x[0])
        print('\nexpected ', y[0].item())
        print('actual   ', y_hat[0].item())
        print()
        break

    average_loss = total_loss.item() / len(test_indices)
    print('Average loss:', average_loss)

if network_output == 'dlon':

    try:
        print(len(hyperparameters))
    except NameError:
        hyperparameters = []
        
    hyperparameters.append((average_loss, {
        'hidden_size': hidden_size,
        'learning_rate': learning_rate,
        'batch_size': batch_size,
        'num_epochs': num_epochs,
        'activation_fcn': activation_fcn,
    }))
```
