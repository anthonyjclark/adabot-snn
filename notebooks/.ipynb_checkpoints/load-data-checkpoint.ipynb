```python
from pathlib import Path
import pandas as pd
```

```python
path_to_csv = Path('../sim/sweep/data1/f0.9/')
csv_files = path_to_csv.glob('*.csv')

df = pd.concat((pd.read_csv(f, comment='#', index_col=False) for f in csv_files), ignore_index=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
# Flip wheel speeds so that positive goes forward
df1 = df.copy()
df1['left'] = -df['left']
df1['right'] = -df['right']
```

```python
df1.head()
```

```python
# Normalize

# Min-Max
df2 = (df1 - df1.min())/(df1.max() - df1.min())

# Standard Score, Standardization, 
# df2 = (df1 - df1.mean())/df1.std()

# df2 = df1

# More info: https://stats.stackexchange.com/questions/7757/data-normalization-and-standardization-in-neural-networks
```

```python
df2.describe()
```

```python
df2.to_pickle(path_to_csv/'pd-data.pickle')
# df.to_pickle(path_to_csv/'pd-data.pickle')
```

```python
pd.unique(df2['left'])
```

```python
df_forward = df2.loc[df2['left'] == df2['right']]
df_forward.describe()
```

```python
df_forward.boxplot(column=['dlon', 'dlat', 'dyaw']);
```

```python
df_rotate = df2.loc[df2['left'] == -df2['right']]
df_rotate.describe()
```

```python
df_rotate.boxplot(column=['dlon', 'dlat', 'dyaw']);
```

```python
diff = 4

x = df2['left'] == (df2['right'] + diff)
y = df2['left'] == (df2['right'] - diff)
z = x | y

df_veer = df2.loc[z]
df_veer.describe()
```

```python
df_veer.boxplot(column=['dlon', 'dlat', 'dyaw']);
```
